```python
# Filter some condtionals.
# Look at the data for people who are making a salary over a certain amount
high_salary = (df['ConvertedComp'] > 70000 )
```

```python
# Narrow data in some criterias
df.loc[high_salary, ['Country', 'LanguageWorkedWith', 'ConvertedComp']]
```

```python
# Make a list of countries that we want to filter out to have less or (|) when we filter something
countries = ['United States', 'India', 'United Kingdom', 'Germany', 'Canada']
filt = df['Country'].isin(countries)
```

```python
df.loc[filt, 'Country']
```

```python
# Get data with 'LanguageWorkedWith'
df['LanguageWorkedWith']
```

```python
# Filter a value in a string
filter_language = df['LanguageWorkedWith'].str.contains('Python', na=False)
```
